2.Explain the difference between mutable and immutable data types with examples?

8.What is a hash table, and how does it relate to dictionaries in Python?
  -Hash Table:
   A data structure that stores key‚Äìvalue pairs. It uses a hash function to convert each key into an index (a number), which decides where the value is stored in memory. This makes lookups, insertions, and deletions very fast (on average O(1) time).

  -Dictionaries in Python:
  Python‚Äôs built-in dict type is implemented using a hash table.
  Keys are hashed to find where to store their values.
  When you do my_dict["name"], Python hashes "name" and quickly jumps to the right location instead of searching through everything.

10.Explain why strings are immutable in Python.
  -Strings are immutable in Python because:

a.Memory & efficiency
Python stores strings in memory in a way that allows them to be shared and reused (string interning).
If strings were mutable, changing one string could unexpectedly affect other variables pointing to the same string.

b.Hashing & dictionary keys
Strings are often used as keys in dictionaries (hash tables).
For hashing to work, the key must never change. Immutability guarantees a string‚Äôs hash value stays constant.

c.Safety & predictability
Since strings are frequently used in programs, making them immutable prevents accidental changes and makes code more reliable.
Example:

s = "hello"
# s[0] = "H"  Error: 'str' object does not support item assignment

# Instead, create a new string
s = "H" + s[1:]
print(s)  # Hello

14.How does the ‚Äúin‚Äù keyword work differently for lists and dictionaries?
  -The in keyword checks for membership, but it behaves differently for lists and dictionaries:

‚úÖ For lists:
in checks if a value exists in the list.
It searches element by element (linear search, O(n) time).
fruits = ["apple", "banana", "cherry"]
print("apple" in fruits)   # True
print("mango" in fruits)   # False

‚úÖ For dictionaries:
in checks if a key exists in the dictionary (not the value).
Uses hashing, so it‚Äôs very fast (O(1) average time).
student = {"name": "Alice", "age": 20}
print("name" in student)     # True  (key check)
print("Alice" in student)    # False (value not checked)

17.Describe the time complexity of accessing elements in a dictionary?
  -Accessing elements in a Python dictionary:
Dictionary lookup is based on hashing.
When you do my_dict[key], Python:
Computes the hash of the key.
Uses that hash to find the index in the underlying hash table.
Retrieves the value directly.

‚è±Ô∏è Time Complexity:
Average case: O(1) ‚Üí constant time (very fast).
Worst case: O(n) ‚Üí happens rarely, only if many keys end up in the same hash bucket (hash collisions).
Example:
person = {"name": "Alice", "age": 25, "city": "Delhi"}
print(person["age"])  # O(1) average time

19.Why are dictionaries considered unordered, and how does that affect data retrieval?
  -A dictionary stores data using a hash table.
Keys are mapped to positions in memory using their hash value, not their insertion position.
Because of this, the order of items isn‚Äôt guaranteed by the data structure itself (at least conceptually).

‚úÖ How this affects data retrieval
Access is by key, not by position
You can‚Äôt reliably say ‚Äúgive me the 2nd item‚Äù like in a list.
Retrieval always uses the key.
person = {"name": "Alice", "age": 25}
print(person["name"])  # ‚úÖ Works
print(person[0])       # ‚ùå Error
